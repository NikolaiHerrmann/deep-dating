```python
import matplotlib.pyplot as plt
import numpy as np
from dating_datasets import load_all_dating_datasets
```

```python
datasets = load_all_dating_datasets()
```

```python
def plot_frequency(ax, dates, title):
    unique_dates, counts = np.unique(dates, return_counts=True)

    ax.bar([str(x) for x in unique_dates], counts)
    ax.set_title(title)

    for tick in ax.get_xticklabels():
        tick.set_rotation(90)

    for bars in ax.containers:
        ax.bar_label(bars)
```

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

for i, dataset in enumerate(datasets):
    plot_frequency(axs[i // 2, i % 2], dataset.img_dates, str(dataset))

fig.supylabel("Number of Images")
fig.supxlabel("Date")
fig.tight_layout()
```
