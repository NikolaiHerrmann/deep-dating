```python
import matplotlib.pyplot as plt
import numpy as np
import os
import glob
import pandas as pd
```

```python
DATASETS_PATH = "../datasets"
```

```python
def plot_frequency(dates):
    unique_dates, counts = np.unique(dates, return_counts=True)
    print("Num images:", np.sum(counts))

    plt.bar([str(x) for x in unique_dates], counts)

    plt.xticks(rotation=90)
    plt.ylabel("Number of Images")
    plt.xlabel("Manuscript Date")

    ax = plt.gca()
    for bars in ax.containers:
        ax.bar_label(bars)

    plt.show()
```

## MPS

```python
def mps_extract_date(img_name):
    return int(os.path.basename(img_name).split("_")[0][3:])

mps_path = os.path.join(DATASETS_PATH, "MPS", "Download", "*", "*")

mps_img_file_names = glob.glob(mps_path)
mps_img_dates = [mps_extract_date(x) for x in mps_img_file_names]

plot_frequency(mps_img_dates)
```
